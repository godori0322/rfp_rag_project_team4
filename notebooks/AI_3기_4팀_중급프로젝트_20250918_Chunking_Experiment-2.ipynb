## Identify Titles, Tables, and Lists

### Subtask:
Analyze the extracted information to identify titles (and their hierarchy), tables, and lists.

**Reasoning**:
Iterate through the extracted page information to identify text blocks that represent titles (based on font size or other characteristics), and potentially identify tables and lists based on layout analysis.

```python
# This is a placeholder for the logic to identify titles, tables, and lists.
# This will require analyzing the 'extracted_info' dictionary.
# The complexity of this step depends on the structure and consistency of the PDF files.
# A simple approach for titles could be based on font size and position.
# Identifying tables and lists is more complex and might require more advanced layout analysis.

identified_elements = {}

for file_name, pages_info in extracted_info.items():
    identified_elements[file_name] = []
    current_headers = {"H1": None, "H2": None, "H3": None}
    for page_info in pages_info:
        # Process blocks to group text more effectively
        for block in page_info.get("blocks", []):
            block_text = ""
            block_type = "text" # Default type

            # Very basic attempt to identify tables - look for multiple spans on the same line
            if len(block.get("lines", [])) > 1:
                first_line = block.get("lines", [])[0]
                if len(first_line.get("spans", [])) > 1:
                     # Check if spans are horizontally aligned (very basic check)
                     span_bboxes = [span.get("bbox") for span in first_line.get("spans", [])]
                     if all(abs(bbox[1] - span_bboxes[0][1]) < 5 for bbox in span_bboxes): # Check y-coordinates are close
                         block_type = "table"

            # Very basic attempt to identify lists - look for common list markers or indentation
            # This is a very naive approach and will not work for most lists
            if block_type == "text": # Only check for list if not identified as table
                 for line in block.get("lines", []):
                      line_text = "".join([span.get("text", "") for span in line.get("spans", [])]).strip()
                      if line_text and (line_text.startswith("- ") or line_text.startswith("* ") or line_text.startswith("• ")):
                           block_type = "list"
                           break # Assume the whole block is a list if one line looks like a list item

            # Concatenate text within the block
            for line in block.get("lines", []):
                for span in line.get("spans", []):
                    block_text += span.get("text", "")

            block_text = block_text.strip()

            if not block_text:
                continue

            # Attempt to identify titles based on the first span's font size in the block
            first_span_size = None
            if block.get("lines"):
                if block.get("lines")[0].get("spans"):
                    first_span_size = block.get("lines")[0].get('spans')[0].get('size') # Corrected line

            is_title = False
            level = None
            if first_span_size and first_span_size > 14: # Example threshold for title font size
                 is_title = True
                 if first_span_size > 18: # Example H1
                     level = "H1"
                 elif first_span_size > 16: # Example H2
                     level = "H2"
                 else: # Example H3
                     level = "H3"

            if is_title:
                 current_headers[level] = block_text
                 # Reset lower level headers
                 if level == "H1":
                     current_headers["H2"] = None
                     current_headers["H3"] = None
                 elif level == "H2":
                      current_headers["H3"] = None

                 identified_elements[file_name].append({
                     "type": "title",
                     "level": level,
                     "text": block_text,
                     "font_size": first_span_size,
                     "bbox": block.get("bbox")
                 })
            else:
                # Add other text as regular content
                 identified_elements[file_name].append({
                    "type": block_type, # Use identified block type (text, table, or list)
                    "text": block_text,
                    "bbox": block.get("bbox"),
                    "headers": current_headers.copy() # Attach current header context
                })


print("Attempted to identify titles, tables, and lists by processing blocks.")
print("Further analysis is needed for accurate identification, especially for complex tables and lists.")
# The identified_elements dictionary now contains identified elements, grouped by block.
# Further processing is needed to accurately identify and classify all elements.
```

## Independent Table/List Chunking

### Subtask:
Extract identified tables and lists as independent chunks.

**Reasoning**:
Iterate through the `identified_elements` and create separate chunks for elements identified as 'table' or 'list'.

```python
# This is a placeholder for independent chunking logic for tables and lists.
# We will now apply a similar chunking approach as the hierarchical one
# to break down large tables or lists if necessary.

def count_tokens(text):
    # A simple way to estimate tokens (e.g., based on words)
    return len(text.split())

independent_chunks = {}
# Define a token limit for independent chunks (can be adjusted)
independent_chunk_token_limit = 2500 # Further Increased limit for independent chunks

for file_name, elements in identified_elements.items():
    independent_chunks[file_name] = []
    for i, element in enumerate(elements):
        if element["type"] in ["table", "list"]:
            element_text = element["text"]
            current_independent_chunk_text = ""
            # Split the text by lines or a similar delimiter for potential chunking
            lines = element_text.splitlines() # Simple split by lines

            for line in lines:
                line_to_add = line.strip()
                if not line_to_add:
                    continue

                # Add the line to the current chunk, preserving some whitespace
                if current_independent_chunk_text:
                    current_independent_chunk_text += "\n" + line_to_add # Add newline between lines
                else:
                    current_independent_chunk_text += line_to_add

                # Check token limit and create a new chunk if exceeded
                if count_tokens(current_independent_chunk_text) > independent_chunk_token_limit:
                    independent_chunks[file_name].append({
                        "text": current_independent_chunk_text,
                        "metadata": {
                            "type": element["type"],
                            "original_element_index": i,
                            "headers": element.get("headers", {})
                        }
                    })
                    current_independent_chunk_text = "" # Start a new chunk

            # Add any remaining text in the current independent chunk
            if current_independent_chunk_text:
                independent_chunks[file_name].append({
                    "text": current_independent_chunk_text,
                    "metadata": {
                        "type": element["type"],
                        "original_element_index": i,
                        "headers": element.get("headers", {})
                    }
                })


print("Attempted to create independent chunks for identified tables and lists, with chunking for larger elements and increased token limit.")
# The independent_chunks dictionary now holds chunks for elements
# that were *very basically* identified as tables or lists, potentially split.
# These chunks are independent of the hierarchical text chunks created earlier.
```

## Hierarchical Chunking

### Subtask:
Create hierarchical chunks based on identified titles and text, adhering to the token length and metadata requirements.

**Reasoning**:
Process the identified elements to create chunks, using the title information to define the hierarchy and attaching the header path as metadata.

```python
# This is a placeholder for the hierarchical chunking logic.
# This will require iterating through 'identified_elements' and grouping
# text based on title changes, keeping track of the current header path.
# We'll also need to consider the 400-800 token length and paragraph integrity.
# A simple token counting approach will be used for now, but for real-world
# scenarios, a proper tokenization library should be used.

def count_tokens(text):
    # A simple way to estimate tokens (e.g., based on words)
    # For more accurate token counting, especially for different languages,
    # consider using libraries like `tokenizers` or `transformers`.
    return len(text.split())

chunked_data = {}

for file_name, elements in identified_elements.items():
    chunked_data[file_name] = []
    current_chunk_text = ""
    current_chunk_headers = {"H1": None, "H2": None, "H3": None}
    current_header_path = ""

    for element in elements:
        if element["type"] == "title":
            # If there's existing text in the current chunk, finalize it before starting a new one
            if current_chunk_text:
                chunked_data[file_name].append({
                    "text": current_chunk_text.strip(),
                    "metadata": {"header_path": current_header_path}
                })
                current_chunk_text = ""

            # Update current headers and header path
            if element["level"] == "H1":
                current_chunk_headers["H1"] = element["text"]
                current_chunk_headers["H2"] = None
                current_chunk_headers["H3"] = None
                current_header_path = element["text"]
            elif element["level"] == "H2":
                current_chunk_headers["H2"] = element["text"]
                current_chunk_headers["H3"] = None
                h1 = current_chunk_headers["H1"] if current_chunk_headers["H1"] else "No H1"
                current_header_path = f"{h1} > {element['text']}"
            elif element["level"] == "H3":
                current_chunk_headers["H3"] = element["text"]
                h1 = current_chunk_headers["H1"] if current_chunk_headers["H1"] else "No H1"
                h2 = current_chunk_headers["H2"] if current_chunk_headers["H2"] else "No H2"
                current_header_path = f"{h1} > {h2} > {element['text']}"

        elif element["type"] == "text":
            text_to_add = element["text"]
            # Add text while trying to preserve whitespace and paragraph integrity
            # Add a space if the current chunk is not empty and the last character is not a space or newline
            if current_chunk_text and not (current_chunk_text.endswith(' ') or current_chunk_text.endswith('\n')):
                current_chunk_text += " " + text_to_add
            else:
                current_chunk_text += text_to_add


            # Basic token length check (using word count as a proxy)
            # Increased the threshold further to reduce the total count significantly.
            if count_tokens(current_chunk_text) > 3000: # Further increased split threshold
                 chunked_data[file_name].append({
                    "text": current_chunk_text.strip(),
                    "metadata": {"header_path": current_header_path}
                 })
                 current_chunk_text = ""

    # Add any remaining text in the current chunk
    if current_chunk_text:
         chunked_data[file_name].append({
             "text": current_chunk_text.strip(),
             "metadata": {"header_path": current_header_path}
         })


print("Attempted hierarchical chunking based on identified titles with further increased chunk size.")
# The chunked_data dictionary now contains chunks based on title hierarchy.
# Further steps are needed to handle tables/lists, refine chunk size, and ensure paragraph integrity.
```

The chunking logic currently splits based on title changes or when a very large token limit is exceeded within a text block. To reduce the chunk count and better adhere to the desired token range (400-800), a more sophisticated chunking strategy is needed that groups smaller text elements semantically, possibly within sections defined by headers, and then splits these larger sections recursively based on token count while trying to preserve paragraph boundaries.
```python
# Refine the chunking process to combine small text elements and apply recursive splitting

def count_tokens(text):
    return len(text.split())

# Define desired chunk token range
min_chunk_tokens = 400
max_chunk_tokens = 800

refined_chunks = {}

for file_name, elements in identified_elements.items():
    refined_chunks[file_name] = []
    current_section_text = ""
    current_section_headers = {"H1": None, "H2": None, "H3": None}
    current_header_path = ""

    def add_section_to_chunks(section_text, metadata):
        """Splits a large section of text into smaller chunks recursively."""
        if not section_text:
            return

        tokens = count_tokens(section_text)

        if tokens <= max_chunk_tokens and tokens >= min_chunk_tokens:
            refined_chunks[file_name].append({"text": section_text.strip(), "metadata": metadata})
        elif tokens < min_chunk_tokens:
             # If a section is too small, add it as a chunk anyway or consider merging in a later step
             refined_chunks[file_name].append({"text": section_text.strip(), "metadata": metadata})
        else: # tokens > max_chunk_tokens
            # Attempt to split the section by paragraphs (double newline)
            paragraphs = section_text.split('\n\n')
            if len(paragraphs) > 1:
                # Recursively chunk paragraphs
                for para in paragraphs:
                    add_section_to_chunks(para.strip(), metadata)
            else:
                # If no paragraphs or still too large, split by lines or a simpler delimiter
                lines = section_text.splitlines()
                temp_chunk = ""
                for line in lines:
                    line_to_add = line.strip()
                    if not line_to_add:
                        continue

                    if temp_chunk and not (temp_chunk.endswith(' ') or temp_chunk.endswith('\n')):
                        temp_chunk += " " + line_to_add
                    else:
                        temp_chunk += line_to_add

                    if count_tokens(temp_chunk) > max_chunk_tokens:
                         refined_chunks[file_name].append({"text": temp_chunk.strip(), "metadata": metadata})
                         temp_chunk = ""
                if temp_chunk:
                    refined_chunks[file_name].append({"text": temp_chunk.strip(), "metadata": metadata})


    for element in elements:
        if element["type"] == "title":
            # Finalize the previous section before starting a new one defined by a title
            add_section_to_chunks(current_section_text, {"header_path": current_header_path, "type": "text"})
            current_section_text = ""

            # Update current headers and header path
            if element["level"] == "H1":
                current_section_headers["H1"] = element["text"]
                current_section_headers["H2"] = None
                current_section_headers["H3"] = None
            elif element["level"] == "H2":
                current_section_headers["H2"] = element["text"]
                current_section_headers["H3"] = None
            elif element["level"] == "H3":
                current_section_headers["H3"] = element["text"]

            h1 = current_section_headers["H1"] if current_section_headers["H1"] else "No H1"
            h2 = current_section_headers["H2"] if current_section_headers["H2"] else "No H2"
            h3 = current_section_headers["H3"] if current_section_headers["H3"] else "No H3"
            current_header_path = f"{h1} > {h2} > {h3}"

        elif element["type"] in ["table", "list"]:
            # Finalize the current text section before adding an independent element
            add_section_to_chunks(current_section_text, {"header_path": current_header_path, "type": "text"})
            current_section_text = ""

            # Add tables and lists as independent chunks (potentially split if very large)
            add_section_to_chunks(element["text"], {"header_path": current_header_path, "type": element["type"]})

        elif element["type"] == "text":
            # Add text to the current section
            text_to_add = element["text"]
            if current_section_text and not (current_section_text.endswith(' ') or current_section_text.endswith('\n')):
                 current_section_text += " " + text_to_add
            else:
                 current_section_text += text_to_add


    # Add any remaining text in the last section
    add_section_to_chunks(current_section_text, {"header_path": current_header_path, "type": "text"})

# Display the refined chunks
print("--- Refined Chunks ---")
for file_name, chunks in refined_chunks.items():
    print(f"File: {file_name}, Number of refined chunks: {len(chunks)}")
    for i, chunk in enumerate(chunks):
        print(f"Chunk {i+1} (Type: {chunk['metadata'].get('type', 'text')})")
        print(f"Header Path: {chunk['metadata'].get('header_path', 'N/A')}")
        print(f"Text (first 200 chars): {chunk['text'][:500]}...")
        print("-" * 20)
    print("\n")

print("Chunking process refined to reduce chunk count and adhere to token limits.")
```

```python
import pdfplumber
import os

def chunk_pdf_file(pdf_path, min_chunk_tokens=300, max_chunk_tokens=800):
    """
    Chunks a PDF file using pdfplumber with structure-aware heuristics.

    This function opens a PDF file, extracts text and layout information
    using pdfplumber, identifies potential titles based on font size and
    position, groups consecutive text blocks, and then splits the text
    into chunks based on specified token limits while attempting to preserve
    paragraph boundaries. It also handles tables and lists as independent
    elements and includes header path metadata in the chunks.

    Args:
        pdf_path: Path to the PDF file.
        min_chunk_tokens: Minimum number of tokens per chunk (default is 300).
        max_chunk_tokens: Maximum number of tokens per chunk (default is 800).

    Returns:
        A list of dictionaries, where each dictionary represents a chunk
        with 'text' and 'metadata' keys. The metadata includes the
        'header_path' (indicating the hierarchical section the chunk belongs to)
        and 'type' (e.g., 'text', 'table', 'list').
    """
    def count_tokens(text):
        """Estimates the number of tokens in a text by counting words."""
        return len(text.split())

    def recursive_chunk_splitter(text, metadata):
        """
        Recursively splits a large section of text into smaller chunks
        based on token limits and paragraph boundaries.
        """
        if not text.strip():
            return []

        tokens = count_tokens(text)
        chunks = []

        if tokens <= max_chunk_tokens and tokens >= min_chunk_tokens:
            chunks.append({"text": text.strip(), "metadata": metadata})
        elif tokens < min_chunk_tokens:
             # If a section is too small, add it as a single chunk
             chunks.append({"text": text.strip(), "metadata": metadata})
        else: # tokens > max_chunk_tokens
            # Attempt to split by paragraphs first
            paragraphs = text.split('\n\n')
            if len(paragraphs) > 1:
                current_paragraph_chunk = ""
                for para in paragraphs:
                    para_to_add = para.strip()
                    if not para_to_add:
                        continue

                    # Check if adding the paragraph exceeds the max token limit
                    if count_tokens(current_paragraph_chunk + "\n\n" + para_to_add) > max_chunk_tokens and current_paragraph_chunk:
                         # If adding the paragraph exceeds the limit and we have existing text, finalize the current chunk
                         chunks.extend(recursive_chunk_splitter(current_paragraph_chunk, metadata)) # Recursively process the current chunk
                         current_paragraph_chunk = para_to_add # Start a new chunk with the current paragraph
                    else:
                        # Otherwise, add the paragraph to the current chunk
                        if current_paragraph_chunk:
                            current_paragraph_chunk += "\n\n" + para_to_add
                        else:
                            current_paragraph_chunk += para_to_add

                # Add any remaining text in the current paragraph chunk
                if current_paragraph_chunk:
                    chunks.extend(recursive_chunk_splitter(current_paragraph_chunk, metadata)) # Recursively process the last chunk

            else:
                # If no clear paragraphs or still too large, split by lines as a fallback
                lines = text.splitlines()
                temp_chunk = ""
                for line in lines:
                    line_to_add = line.strip()
                    if not line_to_add:
                        continue

                    if temp_chunk and not (temp_chunk.endswith(' ') or temp_chunk.endswith('\n')):
                         temp_chunk += " " + line_to_add
                    else:
                         temp_chunk += line_to_add

                    if count_tokens(temp_chunk) > max_chunk_tokens:
                         chunks.append({"text": temp_chunk.strip(), "metadata": metadata})
                         temp_chunk = ""
                if temp_chunk:
                    chunks.append({"text": temp_chunk.strip(), "metadata": metadata})
        return chunks


    doc_elements = [] # Store processed elements (titles and text blocks)
    current_headers = {"H1": None, "H2": None, "H3": None}

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Extract words with bounding boxes and font info
            words = page.extract_words(x_tolerance=1, y_tolerance=1, keep_blank_chars=False, extra_attrs=["size", "fontname"])

            current_text_block = ""
            last_y = None
            last_font_size = None
            last_font_name = None
            is_new_block = True # Flag to check if we are at the beginning of a potential text block

            for word in words:
                text = word['text']
                x0, y0, x1, y1 = word['x0'], word['top'], word['x1'], word['bottom']
                size = word['size']
                fontname = word['fontname']

                # Heuristic for new line/paragraph/block based on vertical distance
                is_new_line = last_y is not None and (y0 - last_y > size * 0.5)
                is_new_paragraph = last_y is not None and (y0 - last_y > size * 1.5)

                # Simple title heuristic based on size and position (left alignment)
                is_potential_title = False
                if is_new_block or is_new_paragraph: # Only check for title at the start of a new block/paragraph
                     if size > 14 and x0 < 100: # Example: larger font and left-aligned
                          is_potential_title = True
                          if last_font_size is not None and size < last_font_size: # Avoid smaller text being identified as title
                              is_potential_title = False


                if is_potential_title:
                     # Finalize current text block before adding a title
                     if current_text_block.strip():
                          doc_elements.append({"type": "text", "text": current_text_block.strip(), "headers": current_headers.copy()})
                     current_text_block = "" # Reset text block

                     # Update headers based on title level (basic size-based heuristic)
                     level = None
                     if size > 18: level = "H1"
                     elif size > 16: level = "H2"
                     else: level = "H3"

                     # Update current headers for subsequent text blocks
                     if level == "H1":
                         current_headers["H1"] = text
                         current_headers["H2"] = None
                         current_headers["H3"] = None
                     elif level == "H2":
                         current_headers["H2"] = text
                         current_headers["H3"] = None
                     elif level == "H3":
                         current_headers["H3"] = text

                     # Add title element (titles themselves are not chunks but define structure)
                     doc_elements.append({"type": "title", "text": text, "level": level})
                     is_new_block = True # Next word starts a new block
                else:
                    # Add text to current block
                    if is_new_paragraph:
                         current_text_block += "\n\n" + text + " "
                    elif is_new_line:
                         current_text_block += "\n" + text + " "
                    else:
                         current_text_block += text + " "
                    is_new_block = False # Not a new block

                last_y = y1
                last_font_size = size
                last_font_name = fontname

            # Add any remaining text block on the page after processing all words
            if current_text_block.strip():
                doc_elements.append({"type": "text", "text": current_text_block.strip(), "headers": current_headers.copy()})


    # Second pass: Chunk based on accumulated elements and headers
    file_chunks = []
    current_chunk_text = ""
    current_chunk_metadata = {"header_path": "N/A", "type": "text"} # Metadata for the current chunk being built

    for element in doc_elements:
        if element["type"] == "title":
            # Finalize current chunk if not empty before a new section defined by a title
            if current_chunk_text.strip():
                file_chunks.extend(recursive_chunk_splitter(current_chunk_text, current_chunk_metadata.copy()))
            current_chunk_text = "" # Reset for new section

            # Update header path for the next chunks based on the new title
            headers = {"H1": None, "H2": None, "H3": None}
            # Find the most recent headers from title elements up to this point
            for prev_element in reversed(doc_elements[:doc_elements.index(element) + 1]):
                 if prev_element["type"] == "title":
                      if prev_element["level"] == "H1" and headers["H1"] is None: headers["H1"] = prev_element["text"]
                      elif prev_element["level"] == "H2" and headers["H2"] is None: headers["H2"] = prev_element["text"]
                      elif prev_element["level"] == "H3" and headers["H3"] is None: headers["H3"] = prev_element["text"]
                 # Stop if we have found at least an H1 or reached the beginning
                 if headers["H1"] is not None or doc_elements.index(prev_element) == 0: break


            h1 = headers["H1"] if headers["H1"] else "No H1"
            h2 = headers["H2"] if headers["H2"] else "No H2"
            h3 = headers["H3"] if headers["H3"] else "No H3"
            current_chunk_metadata["header_path"] = f"{h1} > {h2} > {h3}"
            current_chunk_metadata["type"] = "text" # Next chunks will be text unless specified

        elif element["type"] in ["table", "list"]:
            # Finalize current text chunk before adding an independent element (table/list)
            if current_chunk_text.strip():
                 file_chunks.extend(recursive_chunk_splitter(current_chunk_text, current_chunk_metadata.copy()))
            current_chunk_text = "" # Reset text chunk

            # Add table/list as an independent chunk (apply splitting if very large)
            # Use the header path of the *previous* text chunk or the most recent title
            headers = element.get("headers", {"H1": None, "H2": None, "H3": None})
            h1 = headers["H1"] if headers["H1"] else "No H1"
            h2 = headers["H2"] if headers["H2"] else "No H2"
            h3 = headers["H3"] if headers["H3"] else "No H3"
            header_path = f"{h1} > {h2} > {h3}"
            file_chunks.extend(recursive_chunk_splitter(element["text"], {"header_path": header_path, "type": element["type"]}))


        elif element["type"] == "text":
            # Add text to the current chunk being built
            text_to_add = element["text"]
            if current_chunk_text:
                 current_chunk_text += "\n\n" + text_to_add # Add double newline between text blocks
            else:
                 current_chunk_text += text_to_add


    # Finalize the last chunk after processing all elements
    if current_chunk_text.strip():
         file_chunks.extend(recursive_chunk_splitter(current_chunk_text, current_chunk_metadata.copy()))


    # Post-processing: Merge small text chunks if possible.
    # This pass is to combine small chunks that might have resulted from splitting
    # or from initial small text blocks that didn't meet the min_chunk_tokens.
    final_chunks = []
    temp_merge_chunk_text = ""
    temp_merge_metadata = None

    for i, chunk in enumerate(file_chunks):
        chunk_tokens = count_tokens(chunk["text"])

        if chunk["metadata"]["type"] in ["table", "list"]:
            # If we have pending merged text, add it before the table/list
            if temp_merge_chunk_text.strip():
                 final_chunks.append({"text": temp_merge_chunk_text.strip(), "metadata": temp_merge_metadata.copy()})
                 temp_merge_chunk_text = ""
                 temp_merge_metadata = None
            # Add the independent table/list chunk directly
            final_chunks.append(chunk)
        elif chunk_tokens < min_chunk_tokens:
             # If current chunk is small text, try to merge it
             if temp_merge_chunk_text:
                  # Merge with double newline to preserve paragraph separation if it existed
                  temp_merge_chunk_text += "\n\n" + chunk["text"]
                  # Keep the header path of the first chunk in the merge group
             else:
                  temp_merge_chunk_text = chunk["text"]
                  temp_merge_metadata = chunk["metadata"].copy() # Keep metadata of the first chunk in the merge group

             # Check if merging exceeds max token limit or if it's the last chunk
             if count_tokens(temp_merge_chunk_text) > max_chunk_tokens or i == len(file_chunks) - 1:
                  if temp_merge_chunk_text.strip():
                       final_chunks.append({"text": temp_merge_chunk_text.strip(), "metadata": temp_merge_metadata.copy()})
                  temp_merge_chunk_text = ""
                  temp_merge_metadata = None
        else:
            # If current chunk is large enough text, add it directly
            # If we have pending merged text, add it before this large chunk
            if temp_merge_chunk_text.strip():
                 final_chunks.append({"text": temp_merge_chunk_text.strip(), "metadata": temp_merge_metadata.copy()})
                 temp_merge_chunk_text = ""
                 temp_merge_metadata = None
            # Add the large text chunk
            final_chunks.append(chunk)

    # Add any remaining merged text after the loop
    if temp_merge_chunk_text.strip():
        final_chunks.append({"text": temp_merge_chunk_text.strip(), "metadata": temp_merge_metadata.copy()})


    return final_chunks
```
